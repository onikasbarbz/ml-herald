```python
#Task 1. sum of nested list
def sum_nested_list(nested_list):
    total=0
    for element in nested_list:
        if isinstance(element, list): #checking if the element is list
            total += sum_nested_list(element)
        else:
            total += element
    return total
#sample nested list
nested_list=[1,[2,[3,4],5],6,[7,8]]
result=sum_nested_list(nested_list)
print("Total sum:", result)
            
```

```python
#Task 2. generate all permutations of a string
def generate_permutations(s):
    """
    calculate all possible permutation of a string.This function takes string input.It recursively does depth-first search to generate all posiible outcomes.

    Args:
    string (str): sequence of characters.

    Returns:
    list: a list of all the unique permutations of the char present in a string.
    """
    if len(s)== 1:
        return[s]
    else:
        result =[] # empty list to store possible perm
        for index, char in enumerate(s):
            #creating a new str excluding the current char and recursively get its perm
            for perm in generate_permutations(s[:index]+s[index+1:]):
                #add the current char to the permutation
                result.append(char +perm) 
        return result
#print the list of possible permutations
print("Possible permutations of 'abc' =", generate_permutations('abc'))
print("Possible permutations of 'aab' =", generate_permutations('aab'))
```

```python
#Task 3. Directory Size Calculation
def calculate_directory_size(directory):
    """
    Calculate the total size of directories including all nested files and subdirectories.
    Args:
    directory (dict): a dictionary representing the directory structure.Keys are file or directories name and values are files sizes (in KB) or nested directories representing subdirectories.
    
    Returns:
    int: The total total size of the directory in KB.
    """
    total_size=0
    for key, value in directory.items():
        if isinstance(value, dict): # if value is dictionary, its a subdirectory
            total_size += calculate_directory_size(value)
        else:
            total_size += value #else its a file, add its size
    return total_size
# sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}
print("Total size of the directory:", calculate_directory_size(directory_structure), "KB")
```

```python
#Task 1. COin change problem
def min_coins(coins, amount):
    """
    Finds the minimunm number of coins to makeup a given amount using DP.
    Parameters:
        coins(list[int]): available coin denominations.
        amount (int): Target amount.
    Returns:
        int: Minimum number of coins, or -1 if not possible.
        """
    # initialize DP value with "infinity" for impossible values
    dp= [float('inf')] * (amount +1)
    dp[0] =0 #base case, 0 coins to make amount 0

    #fill the DP table
    for coin in coins:
        for i in range(coin,amount + 1):
            dp[i] =min(dp[i],dp[i- coin]+1)
    #if the target amount is still "infinity," its not possible
    return dp[amount] if dp[amount] !=float('inf') else -1
# test with coins
coins= [1,2,5]
amount= 11

result = min_coins(coins, amount)
print(f"Minimmum coins to make {amount} using {coins}: {result}")
```

```python
#Task 2. Longest Common Subsequence (LCS):
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the longest common subsequence (LCS) between two strings.
    Parameters:
        s1 (str): The first string.
        s2 (str): The second string.
    Returns:
        int: Length of the LCS.
    """
    #create a DP table to store LCS lengths
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    #fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:  #if characters match
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:  #otherwise, take the maximum of the previous subproblems
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    #the bottom-right cell contains the length of the LCS
    return dp[m][n]
#test with s1 = "abcde" and s2 = "ace"
s1 = "abcde"
s2 = "ace"

#calling the function
result = longest_common_subsequence(s1, s2)
print(f"Length of LCS between '{s1}' and '{s2}': {result}")
```

```python
#Task 3.0/1 Knapsack Problem:
def knapsack(weights, values, capacity):
    """
    Finds the maximum value that can be carried in a knapsack with a given capacity.
    This is solved using dynamic programming.
    
    Parameters:
        weights (list[int]): List of weights for each item.
        values (list[int]): List of values for each item.
        capacity (int): The weight capacity of the knapsack.
        
    Returns:
        int: The maximum value that can be obtained within the given capacity.
    """
    n = len(weights)  #number of items
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]  #DP table to store the maximum value at each weight capacity

    # fill the DP table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            #if the item can fit in the current capacity
            if weights[i - 1] <= w:  
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]  #item is excluded if it cannot fit

    return dp[n][capacity]  #the bottom-right cell contains the maximum value
#test with weights = [1, 3, 4, 5], values = [1, 4, 5, 7], and capacity = 7
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

#call the function
result = knapsack(weights, values, capacity)
print(f"Maximum value in the knapsack with capacity {capacity}: {result}")
```
